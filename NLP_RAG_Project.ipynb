## Question Answering using LLM

#### Downloading and Loading the model

```python
#Mistral instruct model seems to be trending on the huggingface website.
#The mistral-7b-instruct-v0.2.Q6_K.gguf provides a model that is 'very large with extremely low quality loss'.
model_name_path = "TheBloke/Mistral-7B-Instruct-v0.2-GGUF"
model_basename = "mistral-7b-instruct-v0.2.Q6_K.gguf"
```

```python
#Downloading the model
model_path = hf_hub_download(
    repo_id=model_name_path,
    filename=model_basename
)
```

```python
#Loading the model with the context set to 10000, batch size set to 512 and gpu layers set to 35
#The maximum sequence length specified on huggingface for this model is 32768, therefore selecting a value less tham this value considering the length of the prompt and response.
#Increased n_ctx to 10000 from 5000 since responses were being truncated.
lcpp_llm = Llama(
    model_path=model_path,
    n_ctx=10000,
    n_batch=512, #Using default batch size, tried others but 512 was giving good results.
    n_gpu_layers=35 #setting to 35 similar to the documentation example for this model on huggingface
)
```

#### Response

```python
# Defining a method to generate response using the above model.
def response(query,max_tokens=500,temperature=0,top_p=0.95,top_k=50):
    model_output = lcpp_llm(
      prompt="[INST]" + query + "[/INST]",
      max_tokens=max_tokens,
      temperature=temperature,
      top_p=top_p,
      top_k=top_k
    )

    return model_output['choices'][0]['text']
```

### System and User Prompt Template

```python
qna_system_message =  """
You are a medical assistant and you need to answer the user question based on your medical knowledge.
The context for this query is provided to you. It begins with the token: ###Context.

User questions will begin with the token: ###Question.

If the answer is not found in the context, respond "I don't know".
"""
```

```python
qna_user_message_template = """
###Context
Here are some documents that are relevant to the question mentioned below.
{context}

###Question
{question}
"""
```

### Response Function

```python
def generate_rag_response(user_input,k=3,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    global qna_system_message,qna_user_message_template
    # Retrieve relevant document chunks
    relevant_document_chunks = retriever.get_relevant_documents(query=user_input,k=k)
    context_list = [d.page_content for d in relevant_document_chunks]

    # Combine document chunks into a single context
    context_for_query = ". ".join(context_list)

    user_message = qna_user_message_template.replace('{context}', context_for_query)
    user_message = user_message.replace('{question}', user_input)

    prompt = "[INST]" + qna_system_message + '\n' + user_message + "[/INST]"

    # Generate the response
    try:
        response = lcpp_llm(
                  prompt=prompt,
                  max_tokens=max_tokens,
                  temperature=temperature,
                  top_p=top_p,
                  top_k=top_k
                  )

        # Extract and print the model's response
        response = response['choices'][0]['text'].strip()
    except Exception as e:
        response = f'Sorry, I encountered the following error: \n {e}'

    return response
```

## Output Evaluation

```python
groundedness_rater_system_message  = """
You are tasked with rating AI generated answers to questions posed by users.
You will be presented a question, context used by the AI system to generate the answer and an AI generated answer to the question.
In the input, the question will begin with ###Question, the context will begin with ###Context while the AI generated answer will begin with ###Answer.

Evaluation criteria:
The task is to judge the extent to which the metric is followed by the answer.
1 - The metric is not followed at all
2 - The metric is followed only to a limited extent
3 - The metric is followed to a good extent
4 - The metric is followed mostly
5 - The metric is followed completely

Metric:
The answer should be derived only from the information presented in the context

Instructions:
1. First write down the steps that are needed to evaluate the answer as per the metric.
2. Give a step-by-step explanation if the answer adheres to the metric considering the question and context as the input.
3. Next, evaluate the extent to which the metric is followed.
4. Use the previous information to rate the answer using the evaluaton criteria and assign a score.
"""
```

```python
relevance_rater_system_message = """
You are tasked with rating AI generated answers to questions posed by users.
You will be presented a question, context used by the AI system to generate the answer and an AI generated answer to the question.
In the input, the question will begin with ###Question, the context will begin with ###Context while the AI generated answer will begin with ###Answer.

Evaluation criteria:
The task is to judge the extent to which the metric is followed by the answer.
1 - The metric is not followed at all
2 - The metric is followed only to a limited extent
3 - The metric is followed to a good extent
4 - The metric is followed mostly
5 - The metric is followed completely

Metric:
Relevance measures how well the answer addresses the main aspects of the question, based on the context.
Consider whether all and only the important aspects are contained in the answer when evaluating relevance.

Instructions:
1. First write down the steps that are needed to evaluate the context as per the metric.
2. Give a step-by-step explanation if the context adheres to the metric considering the question as the input.
3. Next, evaluate the extent to which the metric is followed.
4. Use the previous information to rate the context using the evaluaton criteria and assign a score.
"""
```

```python
user_message_template = """
###Question
{question}

###Context
{context}

###Answer
{answer}

"""
```

```python
def generate_ground_relevance_response(user_input,k=3,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    global qna_system_message,qna_user_message_template
    # Retrieve relevant document chunks
    relevant_document_chunks = retriever.get_relevant_documents(query=user_input,k=3)
    context_list = [d.page_content for d in relevant_document_chunks]
    context_for_query = ". ".join(context_list)

    # Combine user_prompt and system_message to create the prompt
    prompt = f"""[INST]{qna_system_message}\n
                {'user'}: {qna_user_message_template.format(context=context_for_query, question=user_input)}
                [/INST]"""

    print("prompt:", prompt)

    response = lcpp_llm(
            prompt=prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            echo=False
            )

    answer =  response["choices"][0]["text"]

    # Combine user_prompt and system_message to create the prompt
    groundedness_prompt = f"""[INST]{groundedness_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    # Combine user_prompt and system_message to create the prompt
    relevance_prompt = f"""[INST]{relevance_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    response_1 = lcpp_llm(
            prompt=groundedness_prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            echo=False
            )

    response_2 = lcpp_llm(
            prompt=relevance_prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            echo=False
            )

    return response_1['choices'][0]['text'],response_2['choices'][0]['text']
```

```python
#Checking relevance for Q1 with temperature=0.9 and max_tokens=1000
user_input = "What is the protocol for managing sepsis in a critical care unit?"
ground,rel = generate_ground_relevance_response(user_input,temperature=0.9, max_tokens=1000)


print(ground,end="\n\n")
print(rel)
```

**Observation:** With temperature=0.9 and max_tokens=1000 the AI generated response is rated a 5 because it fully adheres to the given metric by being derived solely from the context and not including any additional or external information.

The context is rated only a 3 since the response mentions mentions ICU patient care, supportive care, and monitoring which are related to managing critically ill patients and are somewhat relevant to the question but do not directly address sepsis treatment.

## Actionable Insights and Business Recommendations

* Vector database took a while to create. With increase in the size of the PDF document(s) that need to be loaded, the storage space + the time needed to create the vectorstore will increase.

* Retrieval parameter k is critical as the answer can be spread across multiple contexts. Setting k=8 seemed sufficient for context to be completely relevant for all questions.

* chunk_overlap set to 20 ensure continuity across chunks.

* max_tokens depends on the complexity of the response. Detailed responses need higher values. Simple queries result in concise outputs inspite of large token limits. In this project max_tokens=1000 seemed to satisfy the need of all questions asked.

* Prompt design can influence how the response is formatted and what specific aspects to focus on. Specific questions will generate specific answers.

* The temperature value affects the variability of the response. Setting temperature to 0.5 showed the response quality to be rated a 4. temperature setting of 0.2 or lower yielded responses that were given an evaluation rating of 5.

* top_p=0.5 influences the response quality. Higher values generate responses that are of greater quality. In this scenario top_p of    0.85 or greater generated responses that were given an evaluation rating of 5.

* Default value top_k=50 generated quality responses. There is no need to increase the top_k value in this case.

* The responses when using RAG always started with "Based on the context..." which seems to be something the Mistral model is trained to generate and is hard to get rid off.

In my case, the Mistral model that I used seemed to generate good responses. It seemed to understand the context without RAG. For proprietary information, RAG would be very useful. The temperature, top_p and top_k values can be adjusted based on size and number of documents. The k value for the similarity_search can be adjusted depending on how dispersed the information is in the documents. When information is spread across, k values may need to be larger to generate good responses. chunk_size and overlap can be adjusted to ensure coherence between chunks.
